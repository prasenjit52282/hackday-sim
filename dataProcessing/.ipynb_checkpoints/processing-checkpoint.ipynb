```python
import numpy as np
import pandas as pd
import datetime
```

```python
df=pd.read_csv("./gps.txt",header=None)
df.columns=['lat','long','ault','speed','time','remark']
df=df[['lat','long','time']].copy()
```

```python
df1=pd.read_csv("./label.txt",header=None)
df1.columns=['sT','eT','type']
```

```python
timeStamp=pd.to_datetime(df.time[0],format="%m/%d/%Y %H:%M:%S")
data=\
df1[['sT','eT']].apply(lambda tup:[(timeStamp+datetime.timedelta(minutes=int(tup[0].split(':')[0]),
                                                                seconds=int(tup[0].split(':')[1]))).strftime(format="%m/%d/%Y %H:%M:%S"),
                                  (timeStamp+datetime.timedelta(minutes=int(tup[1].split(':')[0]),
                                                                seconds=int(tup[1].split(':')[1]))).strftime(format="%m/%d/%Y %H:%M:%S")],
                       axis=1).values.tolist()
df2=pd.concat([pd.DataFrame(data,columns=['sT','eT']),df1[['type']]],axis=1)
```

```python
segments=df2.apply(lambda tup:df[(df.time>=tup[0]) & (df.time<=tup[1])][['lat','long']].values.tolist(),axis=1).values.tolist()
```

```python
List=[]

for label, seg in zip(df2.type.values.tolist(),segments):
    seg_list=[]
    for cord in seg:
        seg_list.append({'lat':cord[0],'lng':cord[1]})
    List.append([label,seg_list])
```
